Data Merger

```python
#Environment Setup

import numpy as np
import pandas as pd
import warnings
```

```python
pd.options.display.max_columns = None
warnings.simplefilter(action='ignore')
```

```python
credit_df = pd.read_csv('Data/Loan Data/Cleaned_Data.csv')
credit_df.tail()
```

```python
CPI_df = pd.read_excel('Data/Macroeconomic Data/CPI(U).xlsx')
CPI_df.tail()
```

```python
Inflation_CPI_df = pd.DataFrame(columns=['Year', 'Inflation CPI'])
Inflation_CPI_df['Year'] = CPI_df['Year'].drop(index=[0,9])
```

Inflation Rate = [CPI(n+1) - CPI(n) * 100] / CPI(n)

```python
for i in range(len(CPI_df)-2):
    Inflation_CPI_df['Inflation CPI'].iloc[i] = (CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i+1].sum() - CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i].sum())*100 / CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i].sum()
```

```python
Inflation_CPI_df['Inflation CPI'] = Inflation_CPI_df['Inflation CPI'].astype(np.float64)
```

```python
Inflation_CPI_df
```

```python
GDP_df = pd.read_csv('Data/Macroeconomic Data/GDP.csv')
GDP_df.head()
```

```python
GDP_df = GDP_df.drop([0])
```

```python
GDP_df = pd.DataFrame(GDP_df.iloc[0].transpose())
GDP_df.head()
```

```python
GDP_df = GDP_df.drop(['Line', 'Unnamed: 1'])
```

```python
Year = GDP_df.index
```

```python
GDP_df = pd.DataFrame(GDP_df.values)
```

```python
GDP_df['Year'] = Year
GDP_df = GDP_df.rename(columns={0:'GDP'})
```

```python
GDP_df['Year'] = GDP_df['Year'].astype(np.int64)
GDP_df['GDP'] = GDP_df['GDP'].astype(np.float64)
```

```python
GDP_df.head()
```

```python
USD_df = pd.read_csv('Data/Macroeconomic Data/Nominal_Broad_USD_Index.csv')
USD_df.head()
```

```python
USD_df['DATE'] = USD_df['DATE'].str.split('-').str[0].astype(int)
```

```python
USD_df = USD_df.replace('.', 100)
```

```python
USD_df['DTWEXBGS'] = USD_df['DTWEXBGS'].astype(np.float64)
```

```python
USD_ind_df = USD_df.groupby(['DATE'], as_index=False).mean()
```

```python
USD_ind_df['DATE']
```

```python
USD_ind_df.head()
```

```python
credit_df['issue_d'] = credit_df['issue_d'].str.split('-').str[1]
```

```python
credit_df['issue_d'] = credit_df['issue_d'].apply(lambda x : '20' + x).astype(np.int64)
```

```python
credit_df['Inflation CPI Rate'].iloc[0]
```

```python
credit_df['Inflation CPI Rate'] = pd.Series()
row = 0
for i in credit_df['issue_d']:
    credit_df['Inflation CPI Rate'].iloc[row] = round(float(Inflation_CPI_df[Inflation_CPI_df['Year']==i]['Inflation CPI']),2)
    row +=1
```

```python
credit_df['GDP'] = pd.Series()
row = 0
for i in credit_df['issue_d']:
    credit_df['GDP'].iloc[row] = GDP_df[GDP_df['Year']==i]['GDP']
    row +=1
```

```python
credit_df['USD_Index_Rate'] = pd.Series()
row = 0
for i in credit_df['issue_d']:
    credit_df['USD_Index_Rate'].iloc[row] = USD_ind_df[USD_ind_df['DATE']==i]['DTWEXBGS']
    row +=1
```

```python
credit_df.to_csv("Data/Loan Data/Merged_Data.csv")
```
