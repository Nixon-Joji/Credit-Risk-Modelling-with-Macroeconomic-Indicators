```python
#Environment Setup

import numpy as np
import pandas as pd
import warnings
```

```python
pd.options.display.max_columns = None
warnings.simplefilter(action='ignore')
```

```python
credit_df = pd.read_csv('D:/Off Campus/Projects/Credit Risk Modelling/Data/Loan Data/Cleaned_Data.csv')
credit_df = credit_df.drop('Unnamed: 0', axis=1)
credit_df.head()
```

```python
CPI_df = pd.read_excel('D:/Off Campus/Projects/Credit Risk Modelling/Data/Macroeconomic Data/CPI(U).xlsx')
CPI_df
```

```python
Inflation_CPI_df = pd.DataFrame(columns=['Year', 'Inflation CPI'])
Inflation_CPI_df['Year'] = CPI_df['Year']
```

Inflation Rate = [CPI(n+1) - CPI(n) * 100] / CPI(n)

```python
for i in range(len(CPI_df)-1):
    Inflation_CPI_df['Inflation CPI'].iloc[i] = (CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i+1].sum() - CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i].sum())*100 / CPI_df.drop(columns=['Year', 'HALF1', 'HALF2']).iloc[i].sum()

Inflation_CPI_df['Inflation CPI'] = Inflation_CPI_df['Inflation CPI'].astype(np.float64)
Inflation_CPI_df = Inflation_CPI_df[:-1]
```

```python
Inflation_CPI_df
```

```python
credit_df['issue_d'] = credit_df['issue_d'].str.split('-').str[0]
credit_df['issue_d'] = credit_df['issue_d'].astype(np.int64)
```

```python
credit_df['Inflation CPI Rate'] = credit_df['issue_d'].apply(lambda x: Inflation_CPI_df[Inflation_CPI_df['Year'] == x]['Inflation CPI'].values[0])
```

```python
credit_df.head()
```

```python
credit_df = credit_df.drop(columns=['issue_d'])
```
